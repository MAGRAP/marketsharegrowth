```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import json
import data_parser
from pathlib import Path 
```

```python
URL = "https://www.macrotrends.net/stocks/research"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
results = soup.find_all("a", href= lambda text: "/stocks/industry/" in text.lower())
stock_dict = dict()

for result in results:
    stock_dict[result.text] = ("".join(("https://www.macrotrends.net", result["href"]))).strip()
```

```python
for industry in stock_dict:
    print(">>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>",industry)
    
    URL_industry = stock_dict[industry]
    page = requests.get(URL_industry)
    soup = BeautifulSoup(page.content, "html.parser")
    pattern = re.compile(r"var data = \[(.*?)\]", re.MULTILINE | re.DOTALL)
    scriptParse = re.search(pattern, str(soup))
    
    block = scriptParse.groups()[0]
    res = json.loads("".join(["[", block, "]"]))
    
    dict_company = dict()

    # Get the company asociated with its url
    for company in res:

        link = BeautifulSoup(str(company['link']), "html.parser")
        link = link.find("a")
        link = str(link['href'])
        link = link.replace("stock-price-history", "financial-statements")
        dict_company[company['ticker']] = link
    
    # Get the financial data from each company
    data_dict = dict()

    for key in dict_company.keys():
        
        try:
            # Get the data
            URL = dict_company[key]
            data = data_parser.parser(URL)
            data['ticket'] = key
            data['industry'] = industry
            
            # Export the data
            filepath = Path(f'financial_data/{key}.csv')  
            filepath.parent.mkdir(parents=True, exist_ok=True)  
            data.to_csv(filepath)  
            
            print("============",key, "Loaded")
        except:
            print("============",key, "NOT LOADED")
            continue
```
