```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import json
```

```python
URL = "https://www.macrotrends.net/stocks/research"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("a", href= lambda text: "/stocks/sector/" in text.lower())
stock_dict = dict()

for result in results:
    stock_dict[result.text] = ("".join(("https://www.macrotrends.net", result["href"]))).strip()
   
```

```python
for item in stock_dict.items():
    print(item)
```

```python
sector_1 = 'Finance'
URL_sector = stock_dict[sector_1]
print(URL_sector)
```

```python
page = requests.get(URL_sector)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
#results = soup.find_all('script')[22]
```

```python
pattern = re.compile(r"var data = \[(.*?)\]", re.MULTILINE | re.DOTALL)
```

```python
scriptParse = re.search(pattern, str(soup))
block = scriptParse.groups()[0]
```

```python
res = json.loads("".join(["[", block, "]"]))
```

```python
res
```

```python
dict_industry = dict()

for industry in res:
    
    link = BeautifulSoup(str(industry['link']), "html.parser")
    link = link.find("a")
    link = str(link['href'])
    print(link)
    dict_industry[industry['zacks_x_ind_desc']] = link
```

```python
dict_industry
```

```python
res = json.loads("".join(["[", block, "]"]))
```

```python
res
```

```python
dict_company = dict()

for company in res:
    
    link = BeautifulSoup(str(company['link']), "html.parser")
    link = link.find("a")
    link = str(link['href'])
    link = link.replace("stock-price-history", "income-statement?freq=Q")
    dict_company[company['ticker']] = link
```
