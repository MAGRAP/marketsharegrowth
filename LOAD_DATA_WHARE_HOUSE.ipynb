```python
cursor.execute(
    
"""
DROP TABLE IF EXISTS [FINANCIAL_DWH].[dbo].[FACT_RATIOS]

CREATE TABLE [FINANCIAL_DWH].[dbo].[FACT_RATIOS](
    
    
    [ID_COMPANY] [nchar](10) NULL,
    [Date] [date] NULL,
    
    [Enterprise Value (EV)] [float] NULL,
    [EV / Revenue] [float] NULL,
    [EV / EBITDA] [float] NULL,
    [EV / EBIT] [float] NULL,
    [EV / Invested Capital] [float] NULL,
    [EV / Free Cash FLow] [float] NULL,
    [Free Cash Flow (FCF)] [float] NULL,
    [EV / FCF] [float] NULL,
    [P/E] [float] NULL,
    [P/S] [float] NULL,
    [P/B] [float] NULL,
    [P/CF] [float] NULL,    
)

INSERT INTO [FINANCIAL_DWH].[dbo].[FACT_RATIOS] (

    [ID_COMPANY],
    [Date],
    
    [Enterprise Value (EV)] ,
    [EV / Revenue] ,
    [EV / EBITDA] ,
    [EV / EBIT] ,
    [EV / Invested Capital] ,
    [EV / Free Cash FLow] ,
    [Free Cash Flow (FCF)] ,
    [EV / FCF] ,
    [P/E] ,
    [P/S] ,
    [P/B] ,
    [P/CF]
)
SELECT 

    [ID_COMPANY],
    [Date],

    ROUND(([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand],2) as [Enterprise Value (EV)]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand]) / NULLIF(Revenue, 0), 2) as [EV / Revenue]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand]) / NULLIF(EBITDA, 0), 2) as [EV / EBITDA]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand]) / NULLIF(EBIT, 0), 2) as [EV / EBIT]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand]) / NULLIF([Long-Term Investments], 0), 2) as [EV / Invested Capital]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand]) / NULLIF([Long-Term Investments], 0), 2) as [EV / Free Cash FLow]
    ,ROUND([Cash Flow From Operating Activities] + [Cash Flow From Investing Activities], 2) as [Free Cash Flow (FCF)]
    ,ROUND((([Shares Outstanding] * [Close]) + [Total Liabilities] - [Cash On Hand])/NULLIF([Cash Flow From Operating Activities] + [Cash Flow From Investing Activities], 0), 2) as [EV / FCF]
    ,ROUND([Close] / NULLIF([EPS - Earnings Per Share], 0), 2) as [P/E]
    ,ROUND(([Shares Outstanding] * [Close]) / NULLIF(Revenue, 0), 2) as [P/S]
    ,ROUND([Close] / NULLIF([Book Value Per Share], 0), 2) as [P/B]
    ,ROUND(([Shares Outstanding] * [Close]) / NULLIF([Cash Flow From Operating Activities], 0), 2) as [P/CF]

FROM [FINANCIAL_DWH].[dbo].[FACT_FINANCIAL]
""")

conn.commit()          
```
