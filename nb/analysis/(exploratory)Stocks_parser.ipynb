```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import json
```

```python
URL = "https://www.macrotrends.net/stocks/research"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
results = soup.find_all("a", href= lambda text: "/stocks/industry/" in text.lower())
stock_dict = dict()

for result in results:
    stock_dict[result.text] = ("".join(("https://www.macrotrends.net", result["href"]))).strip()
   
```

```python
for key in stock_dict.keys():
    print(stock_dict[key])
```

```python
industry_1 = 'Large Cap Pharmaceutical'
URL_industry = stock_dict[industry_1]
print(URL_industry)
```

```python
page = requests.get(URL_industry)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
#results = soup.find_all('script')[22]
```

```python
pattern = re.compile(r"var data = \[(.*?)\]", re.MULTILINE | re.DOTALL)
```

```python
scriptParse = re.search(pattern, str(soup))
block = scriptParse.groups()[0]
```

```python
res = json.loads("".join(["[", block, "]"]))
```

```python
res
```

```python
dict_company = dict()

for company in res:
    
    link = BeautifulSoup(str(company['link']), "html.parser")
    link = link.find("a")
    link = str(link['href'])
    link = link.replace("stock-price-history", "financial-statements")
    dict_company[company['ticker']] = link
```

```python
dict_company
```

```python
import data_parser
```

```python
data_dict = dict()

for key in dict_company.keys():
    URL = dict_company[key]
    data = data_parser.parser(URL)
    data_dict[key] = data
```
