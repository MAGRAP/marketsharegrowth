```python
import yfinance as yf
import pandas as pd
import os
from os.path import join
from datetime import datetime
from pathlib import Path 
```

```python
path = 'financial_income_statement_data'
companys = os.listdir(path)
```

```python
companys
```

```python
for company in companys:
    
    try:
        # Get the ticker 
        ticker = company[:-4]

        # Read the csv from stagging
        ticker_file = pd.read_csv(join(path, company))

        # Get the company data from yfinance
        stock_info = yf.Ticker(ticker)
        hist = stock_info.history(period="max", interval = "1mo")

        # Get the price from the closest date from the yfinance data mathcing with the csv date
        list_date = hist.index.astype('str')
        price = list()

        for row in ticker_file['Unnamed: 0']:

            base_date = row

            # Function for parsing strings using specific format
            get_datetime = lambda s: datetime.strptime(s, "%Y-%m-%d")

            base = get_datetime(base_date)
            later = filter(lambda d: get_datetime(d) > base, list_date)
            closest_date = min(later, key = lambda d: get_datetime(d))

            # Append price of the date to the list 
            price.append(hist['Close'].loc[closest_date])

        # Create a dataframe to export the price and date of the company
        historyPrice_df = pd.DataFrame(data = price, index = ticker_file['Unnamed: 0'], columns=['Close'])
        historyPrice_df.index.names = ['Date']
        historyPrice_df['ticker'] = ticker
        
        # Calculate the growth for each date to the next year
        historyPrice_df['Close_1'] = historyPrice_df['Close'].shift(1)
        historyPrice_df['Growth'] = (historyPrice_df['Close_1'] - historyPrice_df['Close']) / historyPrice_df['Close']
        
        # Export the dataframe
        filepath = Path(f'prices_data/{ticker}.csv')  
        filepath.parent.mkdir(parents=True, exist_ok=True)  
        historyPrice_df.to_csv(filepath)
        print(ticker, 'loaded')
        
    except:
        print(ticker, 'not loaded')
        
    
```
