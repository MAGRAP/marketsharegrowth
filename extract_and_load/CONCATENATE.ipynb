```python
import pandas as pd
import os
from os.path import join
from pathlib import Path 
```

```python
def concat_all_files(path):
    
    """
    Given a path, concatenate all files inside
    that path. 
    Files must be in a .csv format
    """
    
    companys = os.listdir(path)

    concatenated_df = pd.DataFrame()

    # Concatenate the financial data
    for company in companys:

        temporal_df = pd.read_csv(join(path, company))
        concatenated_df = pd.concat([concatenated_df, temporal_df])    
        print(company, "loaded")

    return concatenated_df
```

```python
# Concatenate all financial data

path = 'financial_income_statement_data'
allFinancialData_df = concat_all_files(path)

# Export the dataframe
filepath = Path(f'concatenated\concat_financial.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
allFinancialData_df.to_csv(filepath)
```
